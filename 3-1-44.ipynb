\begin{equation}
W_{n + 1} = W_{n} + \eta_{n}(\overrightarrow{x}\overrightarrow{x}^{\mathsf T}W_{n} - W_{n}\Theta_{n}) + O(\eta_{n}^{2})
\end{equation}

### 1. ニューラルネットワークによる次元変換の実装

実験に使用したパラメータを記す.
* $\Theta_{n}$ = $\overrightarrow{y}\overrightarrow{y}^{\mathsf T}$
* $\eta$ = 0.01
* $W_{n}$の初期値は0から1までの一様乱数を格納した縦3横2行列
* 入力 $\overrightarrow{x}$は平均0, 標準偏差1の正規分布をx軸9倍, y軸4倍, z軸にしたもの
* 10000個のデータセットで学習を行う
* 1つのデータに対して行う学習の回数は1000回

```python
ETA = 0.01
XPOWER = 9
YPOWER = 4
ZPOWER = 1
RAND_AVE = 0
RAND_STD = 1

DATA_COUNT = 10000
TRAIN_COUNT = 1000
```

与式の学習に必要な行列計算のプログラムを実装した.

```python
def product(A, B):
    """
    >>a = [[1,2, 3], [4, 5, 6]]
    >>b = [[1, 1], [1, 1], [1, 1]]
    >>print product(a,b) == [[6,6,],[15,15]]
    """
    AH = len(A)
    AW = len(A[0])
    BH = len(B)
    BW = len(B[0])
    if AW != BH:
        print "Size Error"
        print AW, BH
    res = [[0 for x in range(BW)] for y in range(AH)]
    for i in range(AH):
        for j in range(BW):
            for k in range(AW):
                res[i][j] += A[i][k] * B[k][j]
    return res

def transpose(A):
    """
    >>>a = [[1,2], [3,4]]
    >>>transpose(a) == [[1,3],[2,4]]
    True
    """
    AH = len(A)
    AW = len(A[0])
    res = [[0 for x in range(AH)] for y in range(AW)]
    for i in range(AH):
        for j in range(AW):
            res[j][i] = A[i][j]
    return res

def plus(A, B):
    """
    >>>a = [[1,2], [3,4]]
    >>>b = [[1, 0], [0, 1]]
    >>>plus(a,b) == [[2,2],[3,5]]
    True
    """
    AH = len(A)
    AW = len(A[0])
    res = [[0 for x in range(AW)] for y in range(AH)]
    for i in range(AH):
        for j in range(AW):
            res[i][j] = A[i][j] + B[i][j]
    return res

def minus(A, B):
    """
    >>>a = [[1,2], [3,4]]
    >>>b = [[1, 0], [0, 1]]
    >>>minus(a,b) == [[0,2],[3,3]]
    True
    """
    AH = len(A)
    AW = len(A[0])
    res = [[0 for x in range(AW)] for y in range(AH)]
    for i in range(AH):
        for j in range(AW):
            res[i][j] = A[i][j] - B[i][j]
    return res

def times(num, A):
    """
    >>>a = [[1,2], [3,4]]
    >>>times(2,a) == [[2,4],[6,8]]
    True
    """
    AH = len(A)
    AW = len(A[0])
    for i in range(AH):
        for j in range(AW):
            A[i][j] = A[i][j] * num
    return A

def calcX():
    return [
        [XPOWER * random.normalvariate(RAND_AVE, RAND_STD)],
        [YPOWER * random.normalvariate(RAND_AVE, RAND_STD)], 
        [ZPOWER * random.normalvariate(RAND_AVE, RAND_STD)]]

def calcW1():
    return [[random.random() for x in range(2)] for y in range(3)]

def NN(Wn):
    y = product(transpose(Wn), X)
    theata = product(y, transpose(y))
    return plus(Wn, times(ETA, minus(product(X, (product(transpose(X), Wn))), product(Wn, theata)))), y
```

### 3. 実験
10000個のデータセットで学習を行った.  
次元変換されたデータをysという変数で取り出している.

```python
data = [calcX() for x in range(DATA_COUNT)]
Ws = []
ys = []
for X in data:
    W = calcW1()
    for i in range(1000):
        W, y = NN(W)
    Ws.append(W)
    ys.append(y)
```

### 4. 考察

予備実験の出力結果からは, およそ200回の学習で$W_{n}が収束することが確認できる.

10000個のデータセットを次元変換したところ, 出力結果が3次元中で平面をなすように集中したことから, 次元変換が正しく行われていることが確認できる. しかし, 第一主成分, 第二主成分からなる平面と完全には一致しなかった. 一致しない理由として, 主成分分析で計算される理論値と, 近似や収束をもとにした計算機の計算結果の間に誤差が生まれることが考えられる. この誤差を縮める方法として, $\Theta_{n}$の設定をより厳密にすることや, $\Theta_{n}$をはじめとするパラメータを上手に設定することが挙げられる.
